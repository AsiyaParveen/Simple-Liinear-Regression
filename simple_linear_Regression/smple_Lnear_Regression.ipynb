### Simple Linear Regression apply on Banna Data set(Softness)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("banana_ripeness_softness.csv")
```

```python
df.head()
```

```python
x = df[['Days_Ripening']]   
y = df['Softness_Score']    
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
X_train,X_test,Y_Train,Y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_Train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
mse = mean_squared_error(Y_test, y_pred)
mae = mean_absolute_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("MSE:", mse)
print("MAE:", mae)
print("R2 Score:", r2)
```

```python
plt.scatter(x, y, label="Actual Data")
plt.plot(x, model.predict(x), label="Regression Line")
plt.xlabel("Days of Ripening")
plt.ylabel("Softness Score")
plt.title("Banana Ripeness vs Softness (Linear Regression)")
plt.legend()
plt.show()
```
